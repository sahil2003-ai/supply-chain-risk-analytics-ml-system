```python
# ===============================
# IMPORT LIBRARIES
# ===============================
import pandas as pd
import numpy as np
```

```python
# ===============================
# LOAD DATA
# ===============================
file_path = "../data/raw/global_supply_chain_risk_dataset.csv"

df = pd.read_csv(file_path)

print("✅ Data Loaded Successfully\n")
```

```python
# ===============================
# BASIC INFO
# ===============================
print("📊 Shape:", df.shape)
print("\n📌 Columns:\n", df.columns.tolist())
print("\n🔎 Preview:\n", df.head())
```

```python
# ===============================
# CLEAN COLUMN NAMES
# ===============================
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print("\n✅ Cleaned Column Names:\n", df.columns.tolist())
```

```python
# ===============================
# DATE CONVERSION
# ===============================
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

```python
# ===============================
# HANDLE MISSING VALUES
# ===============================
print("\n❗ Missing Values Before:\n", df.isnull().sum())

df = df.dropna()

print("\n✅ Missing Values After:\n", df.isnull().sum())
```

```python
# ===============================
# REMOVE DUPLICATES
# ===============================
df = df.drop_duplicates()
```

```python
# ===============================
# FEATURE ENGINEERING
# ===============================

# Time Features
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# Risk Level
def risk_level(score):
    if score < 3:
        return 'Low'
    elif score < 7:
        return 'Medium'
    else:
        return 'High'

df['risk_level'] = df['geopolitical_risk_score'].apply(risk_level)

# Transport Speed Category
df['transport_category'] = df['transport_mode'].map({
    'Air': 'Fast',
    'Sea': 'Slow',
    'Rail': 'Moderate',
    'Road': 'Variable'
})

# Efficiency Score
df['efficiency_score'] = df['carrier_reliability_score'] / (df['lead_time_days'] + 1)
```

```python
# ===============================
# TARGET VARIABLE CHECK
# ===============================
df['disruption_occurred'] = df['disruption_occurred'].astype(int)
```

```python
# ===============================
# FINAL CHECK
# ===============================
print("\n📊 Final Shape:", df.shape)
print("\n📌 Final Columns:\n", df.columns.tolist())
```

```python
# ===============================
# SAVE CLEAN DATA
# ===============================
output_path = "../data/processed/cleaned_supply_chain.csv"
df.to_csv(output_path, index=False)

print("\n✅ CLEANED DATA SAVED SUCCESSFULLY →", output_path)
```
